```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\soumy\Desktop\Omdena\Startup\Dataset\Startups.csv')
df.head()
```

```python
df.info()
```

```python
# number of missing data points per column
missing_values_count = df.isnull().sum()
missing_values_count
```

```python
#] how many total missing values do we have?
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
#removing unneccesury features
df_req_features = df.drop(['Year Founded','Mapping Location','Office Address','Logo','Seed-DB / Mattermark Profile','Crunchbase / Angel List Profile','Website'],axis=1)
df_req_features.columns
```

```python
contain_values = df_req_features[df_req_features['Company'].str.contains('YC')]
contain_values.head()
```

```python
df_rf=df_req_features[~df_req_features.Company.str.contains('YC')]
df_rf.head()
```

```python
#check if duplicate record is present for any company (no duplicates)
df_rf[df_rf.duplicated(subset=['Company']) == True ]
```

```python
df_rf['Amounts raised in different funding rounds']=df_rf['Amounts raised in different funding rounds'].str.replace('$','')
df_rf['Amounts raised in different funding rounds']=df_rf['Amounts raised in different funding rounds'].str.replace('undisclosed amount','0',)
df_rf['Amounts raised in different funding rounds']=df_rf['Amounts raised in different funding rounds'].fillna(0)
df_rf
```

```python

df2=df_rf['Amounts raised in different funding rounds'].str.split(expand=True)
df2.columns = ['Funding_Round{}'.format(x+1)  for x in df2.columns]
#pd.set_option("display.max_rows", None, "display.max_columns", None)
df2.head()
```

```python
df2.info()
```

```python
# Dropping Round5,6,7,8 as there is very less data
df2.drop(['Funding_Round5','Funding_Round6','Funding_Round7','Funding_Round8'],axis=1,inplace=True)
df2.head()
```

```python

df2['Funding_Round1']=df2['Funding_Round1'].str.replace(',','',)
df2['Funding_Round2']=df2['Funding_Round2'].str.replace(',','',)
df2['Funding_Round3']=df2['Funding_Round3'].str.replace(',','',)
df2['Funding_Round4']=df2['Funding_Round4'].str.replace(',','',)
df2 .head()
```

```python
convert_dict = {'Funding_Round1': float,
                'Funding_Round2': float,
                'Funding_Round3': float,
                'Funding_Round4': float
               }
df2 = df2.astype(convert_dict) 
df2.info()
```

```python
#nonzero_mean = df2[ df2.Funding_Round1 != 0 ].mean
df2 = df2.replace(0, np.NaN)
mean_value=df2.mean()
df2.fillna(value=mean_value, inplace=True)
df2.head()
```

```python
df_rf = df_rf.join(df2)
df_rf.head()
```

```python
df_rf.to_csv("Data_21_Y-Combinator_cleaned_data.csv")
```
