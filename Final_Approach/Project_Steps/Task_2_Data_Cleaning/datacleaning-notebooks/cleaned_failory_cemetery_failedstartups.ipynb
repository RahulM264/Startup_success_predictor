```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("failory_cemetery_failedstartups.csv")
df.head(10)
```

```python
df.info()
```

```python
df.drop(df.index[df['country'] != 'United States'], inplace = True) #dropping the non-USA data
```

```python
# df.drop(df.index[df['total_funding_amount'] == '-' | 'Nothing' | 'No Data'], inplace = True)
df = df[~df['total_funding_amount'].isin(['-','Nothing','No Data']) ] #Excluding invalid/null data from 'total_funding_amount' column
```

```python
df['total_funding_amount'] = df['total_funding_amount'].str.replace('$','') 
```

```python
 '''Converts
 alpha numeric currency amount
 to number'''

def alpha_dollar_to_float_number(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

```

```python
df['total_funding_amount']= df['total_funding_amount'].apply(alpha_dollar_to_float_number)
```

```python
#As the converted strings of range was not in proper data format,converting it manually [As we have only two such values]
df['num_employees'] = df['num_employees'].str.replace('Oct-50','10-50')
df['num_employees'] = df['num_employees'].str.replace('10-Jan','01-10')

df
```

```python
#Replacing '-' with 'to' . Because Pandas converts the string to date formate if it finds '-'.
df['num_employees'] = df['num_employees'].str.replace('-',' to ')
```

```python
df.to_csv("cleaned_failory_cemetery_failedstartups.csv",index=False)
```
