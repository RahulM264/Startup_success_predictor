Cleaning documentation:
1. Filtered for US datasets
2. removed unnecessary and columns with more than 50% missing values
3. if a company was acquired, label it as successful else failed
4. Converted founded date to only year and removed all startups which were founded before 2008.
5. Removed startups with fundings value not in USD to clean the funding column
6. Imputed funding amount with mean
7. Number of employees have 80% missing values and had to dropped. But since it seems like an important feature about whom information is readily available, trying to scrape this info from crunchbase and will be added soon. For now, it has been dropped.

```python
import pandas as pd
import numpy as np
df= pd.read_csv('failure_data.csv')
df.drop(labels= 'Unnamed: 0', axis=1, inplace= True)
df.head()
```

```python
df.dropna(subset=['Headquarters Location'], inplace=True)
df=df[(df.iloc[:]['Headquarters Location'].str.contains("United States"))]
df.head()
```

```python
df.drop(labels= ['Average Visits (6 months)', 'Bounce Rate', 'Bounce Rate Growth', 'CB Rank (Company)', 'Downloads Last 30 Days', 'Founders', 'Global Traffic Rank', 'Headquarters Regions', 'Monthly Rank Change (#)'], axis=1, inplace= True)
df.drop(labels= ['Monthly Rank Growth', 'Monthly Visits', 'Monthly Visits Growth', 'Most Popular Patent Class', 'Most Popular Trademark Class', 'Number of Apps', 'Operating Status', 'Patents Granted', 'Number of Lead Investors', 'Page Views / Visit', 
                 'Page Views / Visit Growth', 'Patents Granted', 'Trademarks Registered', 'Visit Duration', 'Visit Duration Growth', 'Organization Name.1', 'CEO Approval Rating', 'Description2','Estimated Revenue Range', 'Top 5 Investors', 'Website','Exit Date','Owler Estimated Revenue', 'Closed Date', 'Company Type'], axis=1, inplace= True)
df.reset_index(inplace=True)
df.drop(labels= 'index', axis=1, inplace= True)
df.info()
```

```python
df.loc[df['Acquisition Status'] == 'Made Acquisitions', 'success'] = 0
df.loc[df['Acquisition Status'].isnull(), 'success'] = 0
df.loc[df['Acquisition Status'] == 'Made Acquisitions, Was Acquired', 'success'] = 1
df.loc[df['Acquisition Status'] == 'Was Acquired', 'success'] = 1
df['success'].unique()
```

```python
df.drop(labels= ['Acquisition Status', 'Number of Acquisitions', 'Category Groups', 'Number of Competitors', 'Number of Investors', 'Funding Status', 'Total Equity Funding Amount'], axis=1, inplace= True)
df.info()
```

```python
df.dropna(subset=['Organization Name', 'Founded Date', 'Categories'], inplace= True)
df.info()
```

```python
df= df[~df.iloc[:]['Total Funding Amount'].str.contains("£", na=False)]
df= df[~df.iloc[:]['Total Funding Amount'].str.contains("€", na=False)]
df= df[~df.iloc[:]['Total Funding Amount'].str.contains("SGD", na=False)]
```

```python
df['Total Funding Amount'] = df['Total Funding Amount'].apply(lambda x: x.replace('$', '').replace(',', '').replace('£', '') if isinstance(x, str) else x).astype(float)
#for i in range(len(df)):
#  print(type(df.loc[i,'Total Funding Amount']), df.loc[i,'Total Funding Amount'])
```
