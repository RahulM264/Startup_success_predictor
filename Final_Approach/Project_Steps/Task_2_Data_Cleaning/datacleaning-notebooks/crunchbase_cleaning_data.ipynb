```python
#filter only startups based in USA
df1 = df1.loc[df1['country_code'] == 'USA']
df2 = df2.loc[df2['company_country_code'] == 'USA']
df3 = df3.loc[df3['company_country_code'] == 'USA']
df4 = df4.loc[df4['company_country_code'] == 'USA']
```

```python
writer = pd.ExcelWriter('cleaned_crunchbase_data.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
df1.to_excel(writer, sheet_name='Companies')
df2.to_excel(writer, sheet_name='Rounds')
df3.to_excel(writer, sheet_name='Investments')
df4.to_excel(writer, sheet_name='Acquisitions')


# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
