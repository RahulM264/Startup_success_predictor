```python
def unique_values(df):
  
  unique_cities=df["city"].unique()
  unique_cities_len=len(unique_cities)
  print('unique_cities_len: {0} *** unique_cities: {1}'.format(unique_cities_len,unique_cities))
  print('***********************************')
  unique_regions=df["region"].unique()
  unique_regions_len=len(unique_regions)
  print('unique_regions_len: {0} *** unique_regions: {1}'.format(unique_regions_len,unique_regions))
  print('***********************************')
  unique_funding_rounds=df["funding_rounds"].unique()
  unique_funding_rounds_len=len(unique_funding_rounds)
  print('unique_funding_rounds_len: {0} *** unique_funding_rounds: {1}'.format(unique_funding_rounds_len,unique_funding_rounds))
  print('***********************************')
  unique_market=df["market"].unique()
  unique_market_len=len(unique_market) 
  print('unique_market_len: {0} *** unique_market: {1}'.format(unique_market_len,unique_market))
unique_values(df)
```

```python
from collections import Counter
def marketplot():
  df_words=pd.DataFrame(df['market'].value_counts())
  df_words=df_words.rename(columns={'market':'number of startups'})
  df_words['market']=df_words.index
  df_unkown_market=df_words[(df_words['market']=='UNKNOWN') ]
  df_words.drop(df_unkown_market.index, inplace=True)
  df_words=df_words.head(25)
  fig=px.bar(df_words,x="number of startups", y="market",orientation='h',color="number of startups",
            labels={
                "market":"Startup Market",
                "number of startups":"Number Of Startups Per Market"
            },
            title="25 Most Common Markets For USA Startups",
             template='plotly_dark')
  fig.update_layout(
    title_font_family="Times New Roman",
    title_font_color="white")
  fig.show()
marketplot()
```

```python
def regionplot():
  region=pd.DataFrame(df['region'].value_counts())
  region['index1'] = region.index
  df_region=region.head(15)
  region_list=df_region['index1'].tolist()
  fig=px.pie(df_region, values="region", names="index1",
            labels={
                'index1':'Region',
                'region':'Number Of Startups'
            },
            title="Most 15 countries where startups launch",
             template="plotly_dark",
             color_discrete_sequence= px.colors.sequential.Plasma_r)
  fig.update_traces(textposition='inside', textinfo='percent+label')
  fig.update_layout(title_font_color="white")
  fig.show()
regionplot()
```

```python
fig=px.histogram(df, x="funding_rounds", histnorm='percent',nbins=20, 
                 title='In which fundings round most startups get their funds by %',
                 labels={
                     'funding_rounds': 'Funding Rounds'
                 },
                 template='plotly_dark')
fig.update_layout(bargap=0.2)
fig.update_layout(
    title_font_color="white",
    yaxis_title='Percentage of funding'
)
fig.show()
```

```python
def fundingsplot():
  fig = px.scatter(df, x="funding_rounds", y="funding_total_usd")
  fig.update_layout(title="Fundings Total(USD) In Each Round",
                    template='plotly_dark',
                    title_font_color="white",
                    yaxis_title="Fundings Amount(USD)",
                    xaxis_title="Fundings Round")
  fig.show()
fundingsplot()
```

```python
df_rounds=pd.read_csv("/content/cleaned_crunchbase_data - Rounds.csv")
df_rounds.head()
```

```python
def fundingroundtype_Plot():
  df=pd.DataFrame(df_rounds["funding_round_type"].value_counts())
  df=df.rename(columns={'funding_round_type':'counts'})
  df['index1']=df.index
  df=df.head(15)
  fig=px.bar(df,x="counts", y="index1",orientation='h',color="counts",
            labels={
                "counts":"Number Of Startups Being Funded With X Funding Type",
                "index1":"Funding Round Type"
            },
            title="15 Most Common Funding Types For USA Startups",
            template='plotly_dark')
  fig.update_layout(
    title_font_family="Times New Roman",
    title_font_color="white")
  fig.show()
fundingroundtype_Plot()
```

```python
def fundings_type_plot():
  fig = px.violin(df_rounds, y=, x="funding_round_type", box=True)
  fig.update_layout(title="Raised Amount Of Fundings(USD) For Each Funding Type",
                    template='plotly_dark',
                    title_font_color="white",
                    yaxis_title="Raised Amount(USD)",
                    xaxis_title="Fundings Round Type")
  fig.show()
fundings_type_plot()
```
