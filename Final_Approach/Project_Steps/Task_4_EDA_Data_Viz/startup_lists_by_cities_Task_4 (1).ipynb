**Data Visualization**

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = STOPWORDS
wordcloud = WordCloud(width=800,height=400,stopwords=stopwords, min_font_size=10,max_words=150).generate(' '.join
                                                                                           (data['description']))
```

```python
plt.figure(figsize=(20,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("Wordcloud of description", fontsize=25)
plt.savefig('1.jpeg')
plt.show()
```

*Word Cloud also called tag cloud is a Visual representation of text data. above word cloud show us the Description of our data*





```python
wordcloud1 = WordCloud(width=800,height=400,stopwords=stopwords, min_font_size=10,max_words=150).generate(' '.join
                                                                                           (data['city']))
                            
```

```python
plt.figure(figsize=(20,8))
plt.imshow(wordcloud1)
plt.axis('off')
plt.title("Wordcloud of state", fontsize=25)
plt.savefig('2.jpeg')
plt.show()
```

*Here Word Clouds show us Difference city's of  State*

```python
f,ax = plt.subplots(figsize =(20,10))
sns.countplot(data['city'],ax=ax)
plt.xticks(rotation=90)
plt.show()
```
